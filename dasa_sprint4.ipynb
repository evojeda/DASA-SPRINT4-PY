```python
#DADOS FICTÍCIO
demanda = [4, 6, 3, 5]      #consumo diário (exemplo)
estoque_inicial = 10        #estoque no primeiro dia
repos_max = 2
repos_max = 8               #limite padrão de reposição por dia
estoque_ideal = 10          #nível alvo
c_falta = 10                #custo por faltar insumo
c_excesso = 2               #custo por exceder o ideal
c_pedido = 1.0              #custo por unidade reposta
c_estoque = 0.5             #custo por unidade armazenada ao fim do dia

#RECURSIVA
def custo_recursivo(dia, estoque):
    if dia == len(demanda):
        return 0
    repor = 0
    consumo = demanda[dia]
    novo_estoque = estoque - consumo + repor
    falta = 0
    excesso = 0
    if novo_estoque < 0:
        falta = (-novo_estoque) * c_falta
        novo_estoque = 0
    elif novo_estoque > estoque_ideal:
        excesso = (novo_estoque - estoque_ideal) * c_excesso
    custo_op = c_pedido * repor + c_estoque * novo_estoque
    melhor = falta + excesso + custo_op + custo_recursivo(dia + 1, novo_estoque)
    for repor in range(1, repos_max + 1):
        consumo = demanda[dia]
        novo_estoque = estoque - consumo + repor
        falta = 0
        excesso = 0
        if novo_estoque < 0:
            falta = (-novo_estoque) * c_falta
            novo_estoque = 0
        elif novo_estoque > estoque_ideal:
            excesso = (novo_estoque - estoque_ideal) * c_excesso
        custo_op = c_pedido * repor + c_estoque * novo_estoque
        custo_total = falta + excesso + custo_op + custo_recursivo(dia + 1, novo_estoque)
        if custo_total < melhor:
            melhor = custo_total
    return melhor

#MEMOIZAÇÃO
memo = {}
def custo_memo(dia, estoque):
    if dia == len(demanda):
        return 0
    if (dia, estoque) in memo:
        return memo[(dia, estoque)]
    repor = 0
    consumo = demanda[dia]
    novo_estoque = estoque - consumo + repor
    falta = 0
    excesso = 0
    if novo_estoque < 0:
        falta = (-novo_estoque) * c_falta
        novo_estoque = 0
    elif novo_estoque > estoque_ideal:
        excesso = (novo_estoque - estoque_ideal) * c_excesso
    custo_op = c_pedido * repor + c_estoque * novo_estoque
    melhor = falta + excesso + custo_op + custo_memo(dia + 1, novo_estoque)
    for repor in range(1, repos_max + 1):
        consumo = demanda[dia]
        novo_estoque = estoque - consumo + repor
        falta = 0
        excesso = 0
        if novo_estoque < 0:
            falta = (-novo_estoque) * c_falta
            novo_estoque = 0
        elif novo_estoque > estoque_ideal:
            excesso = (novo_estoque - estoque_ideal) * c_excesso
        custo_op = c_pedido * repor + c_estoque * novo_estoque
        custo_total = falta + excesso + custo_op + custo_memo(dia + 1, novo_estoque)
        if custo_total < melhor:
            melhor = custo_total
    memo[(dia, estoque)] = melhor
    return melhor

#ITERATIVA (BOTTOM-UP)
n = len(demanda)
max_estoque = 30
dp = [[0] * (max_estoque + 1) for _ in range(n + 1)]
for e in range(max_estoque + 1):
    dp[n][e] = 0
for dia in range(n - 1, -1, -1):
    for estoque in range(max_estoque + 1):
        repor = 0
        novo_estoque = estoque - demanda[dia] + repor
        falta = 0
        excesso = 0
        if novo_estoque < 0:
            falta = (-novo_estoque) * c_falta
            novo_estoque = 0
        elif novo_estoque > estoque_ideal:
            excesso = (novo_estoque - estoque_ideal) * c_excesso
        if novo_estoque > max_estoque:
            novo_estoque = max_estoque
        custo_op = c_pedido * repor + c_estoque * novo_estoque
        melhor = falta + excesso + custo_op + dp[dia + 1][novo_estoque]
        for repor in range(1, repos_max + 1):
            novo_estoque = estoque - demanda[dia] + repor
            falta = 0
            excesso = 0
            if novo_estoque < 0:
                falta = (-novo_estoque) * c_falta
                novo_estoque = 0
            elif novo_estoque > estoque_ideal:
                excesso = (novo_estoque - estoque_ideal) * c_excesso
            if novo_estoque > max_estoque:
                novo_estoque = max_estoque
            custo_op = c_pedido * repor + c_estoque * novo_estoque
            custo_total = falta + excesso + custo_op + dp[dia + 1][novo_estoque]
            if custo_total < melhor:
                melhor = custo_total
        dp[dia][estoque] = melhor

# RESULTADOS
c1 = custo_recursivo(0, estoque_inicial)
c2 = custo_memo(0, estoque_inicial)
c3 = dp[0][min(estoque_inicial, max_estoque)]

print("Custo mínimo (Recursivo):   ", c1)
print("Custo mínimo (Memoização):  ", c2)
print("Custo mínimo (Iterativa):   ", c3)
```
